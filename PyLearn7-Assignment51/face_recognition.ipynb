# Generating dataset:

```python
def generate_dataset(path):
    embedding_obj_features = DeepFace.represent(img_path= path , model_name="ArcFace" , enforce_detection=False)
    return embedding_obj_features


dic = defaultdict(list)

for i in range(len(folders_names)) :

    print(folders_names[i])
    folder_path = folders_paths[i]
    print(folder_path)

    each_folder_files = []
    for file_path in os.listdir(folder_path) :
        each_folder_files.append(file_path)
        complete_path = []

    for file in each_folder_files:
        path = folder_path + "/" + file
        complete_path.append(path)

    for j in range(len(each_folder_files)) :
        face_feature_numbers = generate_dataset(complete_path[j])
        dic['Label'].append(folders_names[i])
        dic['File_name'].append(each_folder_files[j] )

        for k in range(len(face_feature_numbers[0]["embedding"])) :
            dic[f'Feature{k+1}'].append(face_feature_numbers[0]["embedding"][k])

    print(i)
```

```python
df = pd.DataFrame(dic)
df
```

```python
df['label_num'] = 0

for i in range(len(df)):
    for j in range(len(folders_names)):

        if df['Label'][i] == folders_names[j]:
            df['label_num'][i] = j

df
```

```python
df.to_csv('/content/drive/MyDrive/Colab Notebooks/PyLearn7-Assignment51/datasets/PersianFace_dataset.csv', index=False)
```

# Machine learning (MLP):

```python
from sklearn.model_selection import train_test_split
import numpy as np
import tensorflow as tf
```

```python
df.iloc[: , 2:-1]
```

```python
X_train = np.array(df.iloc[: , 2:-1])
Y_train = np.array(df["label_num"])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_train, Y_train, test_size=0.2)
Y_test = Y_test.reshape(-1 , 1)
Y_train = Y_train.reshape(-1 , 1)
X_train = X_train / 255
X_test = X_test / 255
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(512),                           # input layer
    tf.keras.layers.Dense(256, activation="relu"),        # hidden layer 1
    tf.keras.layers.Dense(128, activation="relu"),        # hidden layer 2
    tf.keras.layers.Dense(30, activation="softmax")       # output layer
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adamax(),
              loss=tf.keras.losses.sparse_categorical_crossentropy,
              metrics=['accuracy'])
```

```python
output_model = model.fit(X_train, Y_train, epochs=200)
loss , accuracy = model.evaluate(X_test , Y_test )
print("Test loss:" , loss)
print("Test accuracy:" , accuracy)
```

```python
model.save('/content/drive/MyDrive/Colab Notebooks/PyLearn7-Assignment51/weights/face_recognition_model.h5')
```

```python
import matplotlib.pyplot as plt

plt.plot(output_model.history["loss"])
plt.plot(output_model.history["accuracy"])

plt.title("Training loss & accuracy")
plt.xlabel("Epoches")
plt.legend(["loss" , "accuracy"])
plt.show()
```
