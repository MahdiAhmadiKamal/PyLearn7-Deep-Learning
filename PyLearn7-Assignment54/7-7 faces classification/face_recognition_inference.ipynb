```python
import tensorflow as tf
import numpy as np
import cv2
```

```python
model = tf.keras.models.load_model('weights/7-7-faces_model.keras')
```

```python
labels = ['Ali_Khamenei', 'Angelina_Jolie', 'Barak_Obama', 'Behnam_Bani', 'Donald_Trump', 'Emma_Watson', 'Han_Hye_Jin', 'Kim_Jong_Un', 'Leyla_Hatami', 'Lionel_Messi', 'Michelle_Obama', 'Morgan_Freeman', 'Queen_Elizabeth', 'Scarlett_Johansson']

image = cv2.imread("7-7-faces/Han_Hye_Jin/Han-Hye-Jin-100_01.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (224, 224))
image = image / 255.0
image = image.reshape(1, 224, 224, 3)

output = model.predict(image)
print(output)
```

```python
print(np.argmax(output))
print(labels[np.argmax(output)])
```
