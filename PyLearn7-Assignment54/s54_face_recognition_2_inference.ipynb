```python
import tensorflow as tf
import numpy as np
import cv2
```

```python
model = tf.keras.models.load_model('/content/drive/MyDrive/Colab Notebooks/PyLearn7-session54/my_7-7-faces_model.keras')
```

```python
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/PyLearn7-session54/7-7-faces/Han_Hye_Jin/Han-Hye-Jin-100_01.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (224, 224))
image = image / 255.0
image = image.reshape(1, 224, 224, 3)

output = model.predict(image)
print(output)
```
