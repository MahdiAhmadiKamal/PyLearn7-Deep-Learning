```python
labels = ['clergyman', 'fellow']

image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/PyLearn7-Assignment54/fellow and clergyman/fellow-and-clergyman/fellow/016.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (224, 224))
image = image / 255.0
image = image.reshape(1, 224, 224, 3)

output = model.predict(image)
print(output)
```
